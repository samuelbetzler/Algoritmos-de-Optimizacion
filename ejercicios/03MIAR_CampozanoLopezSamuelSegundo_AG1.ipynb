## Divide y vencerás - Problema Torres de Hanói
- Parto de un problema y lo soluciono mediante la solución de subproblemas y unión de resultados
- En este caso concreto: H(n) = H(n-1) + 1 + H(n-1)
- Pensar lo anterior como... "Muevo todos los discos salvo uno a torre auxiliar, muevo el disco más grande a la torre final y muevo todos los discos restantes de la torre auxiliar a la final"
- Solamente resuelvo el problema de mover un disco (condición de parada de división) porque el resto se aplica el mismo algoritmo.
- Mover un disco es muy sencillo -> Solución directa -> Mover de torre de incio a torre de fin
- Ojo, el print del "if" es mover un disco entre torres (cualesquiera) y el print del "else" es el caso de mover el disco grande a torre final

```python
def torres_hannoi(numero_discos, inicio, fin):

    #print(f"torres_hannoi({numero_discos}, {inicio}, {fin})")

    # Si el número de discos es 1 -> Caso base, mover de inicio a fin
    if numero_discos == 1:
        #print(f"Caso base: Mover desde {inicio} a {fin} -- torres_hannoi({numero_discos}, {inicio}, {fin})")
        print(f"Mover desde {inicio} a {fin}")
    # Si el número de discos es mayor que 1 -> Mover resto de discos, mover disco único,
    # recuperar => Recursivamente
    else:
        torres_hannoi(numero_discos-1, inicio, 6-inicio-fin)
        #print(f"Caso central: Mover desde {inicio} a {fin} -- torres_hannoi({numero_discos}, {inicio}, {fin})")
        print(f"Mover desde {inicio} a {fin}")
        torres_hannoi(numero_discos-1, 6-inicio-fin, fin)

torres_hannoi(3,1,3)
```

## Técnicas voraces - Problema del cambio
- Para resolver el problema, elijo una solución "buena" sin pensar en consecuencias
- Sería como elegir algo "óptimo" localmente con la esperanza de que sea lo mejor en general
- No siempre funciona bien -> Si en el conjunto de las soluciones, dos están próximas entre sí y generan una solución mayor puede fallar (ej: monedas [1, 3, 4, 9])

```python
def cambio_monedas(sistema, cambio):

    print(f"Sistema: {sistema}")
    print(f"Cambio: {cambio}")

    # Inicialización de resultados
    resultado = [0] * len(sistema)

    # Para cada posible moneda, calculamos el número de monedas máximo que podemos dar
    # y lo restamos del cambio pendiente (mejor decisión local, que no tiene por qué
    # ser óptima
    for i in range(len(sistema)):
        total_monedas = cambio//sistema[i] # División con resultado entero
        resultado[i] = total_monedas
        cambio = cambio - resultado[i]*sistema[i]

    return resultado

# Sistema óptmo
sistema = [25, 10, 5, 1]
cambio = 27
resultado = cambio_monedas(sistema,cambio)
print(f"Resultado: {resultado}")

print("")

# Sistema NO óptimo
sistema = [11, 5, 1]
cambio = 15
resultado = cambio_monedas(sistema,cambio)
print(f"Resultado: {resultado}")
```

## Backtracking - Problema de las N reinas
- Se divide el problema en etapas
- En cada etapa se construye una solución parcial
- Se comprueba si esa solución parcial sigue siendo válida
- Si es válida, se continúa con la siguiente etapa
- Si no es válida, se descarta y se prueba otra alternativa
- Si ninguna alternativa funciona, se retrocede a la etapa anterior para cambiar decisiones

```python
# Función que comprueba si por columnas hay un error en la solución
# etapa(i) = etapa (j) -> Error
# También comprueba si por diagonales hay un error en la solución
# (abs(i-j) = abs(xi-xj)) -> Error
def conflicto(solucion, etapa):
    for i in range(etapa+1):
        if solucion.count(solucion[i]) > 1:
            return True
        for j in range(etapa):
            if (i != j):
                if (abs(i-j) == abs(solucion[i]-solucion[j])):
                    return True
    else:
        return False
```

```python
# Búsqueda de una solución
def colocar_reinas(numero_reinas, solucion = None, etapa = 0):

    if solucion is None:
        solucion = numero_reinas*[0]

    # Caso base: si he colocado todas las reinas (he completado todas las etapas)
    # etapa comienza en 0 y llega hasta numero_reinas
    if numero_reinas == etapa:
        #print(f"({numero_reinas}, {solucion}, {etapa})")
        print(f"{solucion}")
        return solucion

    else:
        # Recorro todas las columnas posibles para la reina en la fila (etapa)
        for i in range(1,numero_reinas+1):
            solucion[etapa] = i
            #print(f"({numero_reinas}, {solucion}, {etapa})")

            # Compruebo si la colocación actual no entra en conflicto
            # con reinas anteriores (columnas y diagonales)
            if not (conflicto(solucion, etapa)):
                # Si es válida, intento resolver el resto del tablero
                resultado = colocar_reinas(numero_reinas, solucion, etapa + 1)

                # Si desde aquí se encuentra una solución válida,
                # la propago hacia arriba para detener la búsqueda
                # En otras palabras, ya tengo resultado, no necesito
                # seguir probando, lo devuelvo
                if resultado is not None:
                    return resultado

        # Limpio la posición de la reina si todos los estados fallan (vuelta atrás)
        solucion[etapa] = 0

        # Si ninguna columna funciona en esta etapa, esta rama no tiene solución
        #return None

solucion = colocar_reinas(4)
```

## 1. Problema en 1D (Lista de Números)

Dado un conjunto de números reales (puntos en 1D), queremos encontrar el par con la **menor distancia absoluta**.

### Primer Intento: Fuerza Bruta
Consiste en comparar cada elemento con todos los demás.
- **Complejidad Temporal:** $O(N^2)$ (doble bucle anidado).
- **Complejidad Espacial:** $O(1)$.
- **¿Se puede mejorar?** Sí, considerablemente.

### Segundo Intento: Ordenación y Escaneo Lineal
Si ordenamos la lista, los puntos más cercanos deben ser **adyacentes** en la lista ordenada.
- Ordenar el array en orden ascendente: $O(N \log N)$.
- Escanear comparando solo elementos contiguos: $O(N)$.
- **Complejidad Temporal Total:** $O(N \log N)$.
- **Complejidad Espacial:** $O(1)$ (o $O(N)$ si se duplica la estructura al ordenar).

### Tercer Intento: Divide y Vencerás
Aunque el método anterior es el más sencillo y eficiente para 1D, el enfoque de Divide y Vencerás funciona así:
1. Ordenamos los puntos de menor a mayor.
2. Dividimos el conjunto en dos mitades (izquierda y derecha).
3. Buscamos recursivamente el par más cercano en cada mitad ($d_L$ y $d_R$).
4. Definimos $d = \min(d_L, d_R)$.
5. **Paso de Combinación:** el único par que puede cruzar la división y estar a distancia menor que $d$ es el formado por el último elemento de la izquierda y el primero de la derecha. Se calcula en $O(1)$.

- **Relación de Recurrencia:** $T(N) = 2\,T(N/2) + O(1)$, equivalente a $O(N)$ una vez ordenado.
- **Complejidad Temporal Total:** $O(N \log N)$ (debido al ordenamiento inicial).

## 2. Extensión a 2D (Coordenadas Cartesianas)

Dada una lista de puntos en el plano: $[(x_1, y_1), (x_2, y_2), \dots]$.

### Fuerza Bruta 2D
Compara la distancia euclidiana de todos los pares.
- **Complejidad Temporal:** $O(N^2)$.

### Divide y Vencerás 2D
Para lograr una complejidad de $O(N \log N)$:
1. Ordenamos los puntos por su coordenada X ($P_x$) y por su coordenada Y ($P_y$).
2. Dividimos verticalmente el conjunto en dos partes iguales usando la mediana en X.
3. Encontramos la distancia mínima recursivamente en ambas mitades: $d = \min(d_L, d_R)$.
4. **Paso de Combinación:**
   - Filtramos los puntos dentro de una franja vertical de ancho $2d$ alrededor de la línea de división.
      - Como ya tenemos la lista ordenada por Y ($P_y$), construimos la franja ordenada por Y en tiempo lineal $O(N)$.
         - Para cada punto, solo comparamos con los siguientes puntos cuya distancia vertical sea menor que $d$. En el plano hay un **máximo de 7 puntos** posibles por cada punto analizado.
            - Así, el paso de combinación se realiza en $O(N)$ comparaciones.

            - **Relación de Recurrencia:** $T(N) = 2\,T(N/2) + O(N)$, lo que resulta en $O(N \log N)$.

            ---

            ## 3. Extensión a 3D (Coordenadas en el Espacio)

            Dada una lista de puntos en el espacio: $[(x_1, y_1, z_1), (x_2, y_2, z_2), \dots]$.

            ### Fuerza Bruta 3D
            Compara la distancia euclidiana 3D de todos los pares.
            - **Complejidad Temporal:** $O(N^2)$.

            ### Divide y Vencerás 3D
            El algoritmo se extiende de forma natural a tres dimensiones manteniendo $O(N \log N)$:
            1. Pre-ordenamos los puntos por X e Y.
            2. Dividimos el espacio con un plano vertical $X = \text{mediana}$.
            3. Recursivamente obtenemos $d = \min(d_L, d_R)$ de ambas mitades.
            4. **Paso de Combinación:**
               - Filtramos los puntos en una franja 3D de ancho $2d$ alrededor del plano de corte ($|p.x - \text{mediana}.x| < d$).
                  - Mantenemos los puntos ordenados por Y (en tiempo $O(N)$ filtrando desde la lista pre-ordenada en Y).
                     - Para cada punto, recorremos los sucesores en la lista ordenada por Y y nos detenemos en cuanto $\Delta Y \ge d$.
                        - En 3D, el volumen que puede contener puntos a distancia menor que $d$ está acotado por una constante (**máximo 15 puntos** por esfera de empaquetamiento). El bucle interno se ejecuta en tiempo amortizado $O(1)$.

                        - **Relación de Recurrencia:** $T(N) = 2\,T(N/2) + O(N)$, garantizando $O(N \log N)$.

```python
# =============================================================================
#  IMPORTS Y UTILIDADES COMUNES
# =============================================================================
#  Importamos las librerias estandar necesarias para el proyecto:
#    - math   : calculo de raices cuadradas (distancia euclidiana).
#    - random : generacion de conjuntos de puntos aleatorios para las pruebas.
#    - time   : medicion de tiempos de ejecucion (benchmark).
#    - typing : anotaciones de tipos para un codigo mas legible y profesional.
# =============================================================================

import math
import random
import time
from typing import List, Tuple, Sequence


# -----------------------------------------------------------------------------
#  Funciones de distancia
# -----------------------------------------------------------------------------
def distancia_1d(a: float, b: float) -> float:
    """Distancia absoluta entre dos numeros (puntos en 1D)."""
    return abs(a - b)


def distancia_euclidiana(p: Sequence[float], q: Sequence[float]) -> float:
    """
    Distancia euclidiana entre dos puntos de cualquier dimension (2D, 3D, ...).

    Parametros
    ----------
    p, q : secuencias de coordenadas (ej. (x, y) en 2D o (x, y, z) en 3D).

    Retorna
    -------
    float : la distancia euclidiana sqrt(sum((p_i - q_i)^2)).
    """
    return math.sqrt(sum((pi - qi) ** 2 for pi, qi in zip(p, q)))
```

## Implementación en 1D

A continuación se implementan los **tres enfoques** para encontrar el par de puntos más cercanos en una lista de números:

1. **Fuerza Bruta** — $O(N^2)$
2. **Ordenación + Escaneo Lineal** — $O(N \log N)$
3. **Divide y Vencerás** — $O(N \log N)$

Cada función devuelve una tupla `(distancia_minima, (punto_a, punto_b))`.

```python
# =============================================================================
#  PROBLEMA EN 1D : Par de puntos mas cercanos en una lista de numeros
# =============================================================================

# -----------------------------------------------------------------------------
#  1) FUERZA BRUTA  ->  O(N^2)
# -----------------------------------------------------------------------------
def mas_cercanos_1d_fuerza_bruta(puntos: List[float]) -> Tuple[float, Tuple[float, float]]:
    """
    Compara cada punto con todos los demas (doble bucle anidado).

    Complejidad temporal : O(N^2)
    Complejidad espacial : O(1)
    """
    n = len(puntos)
    mejor_dist = float('inf')
    mejor_par = (None, None)

    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_1d(puntos[i], puntos[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (puntos[i], puntos[j])

    return mejor_dist, mejor_par


# -----------------------------------------------------------------------------
#  2) ORDENACION + ESCANEO LINEAL  ->  O(N log N)
# -----------------------------------------------------------------------------
def mas_cercanos_1d_orden_scan(puntos: List[float]) -> Tuple[float, Tuple[float, float]]:
    """
    Una vez ordenada la lista, el par mas cercano SIEMPRE es adyacente.
    Por tanto basta con recorrer comparando elementos contiguos.

    Complejidad temporal : O(N log N)  (dominada por la ordenacion)
    Complejidad espacial : O(N)        (lista ordenada)
    """
    ordenados = sorted(puntos)          # O(N log N)
    mejor_dist = float('inf')
    mejor_par = (None, None)

    for i in range(len(ordenados) - 1):  # O(N)
        d = distancia_1d(ordenados[i], ordenados[i + 1])
        if d < mejor_dist:
            mejor_dist = d
            mejor_par = (ordenados[i], ordenados[i + 1])

    return mejor_dist, mejor_par


# -----------------------------------------------------------------------------
#  3) DIVIDE Y VENCERAS  ->  O(N log N)
# -----------------------------------------------------------------------------
def mas_cercanos_1d_divide_venceras(puntos: List[float]) -> Tuple[float, Tuple[float, float]]:
    """
    Enfoque recursivo Divide y Venceras:
      1. Se ordenan los puntos (una sola vez).
      2. Se divide la lista en dos mitades.
      3. Se resuelve cada mitad recursivamente -> dL, dR.
      4. Combinacion: el unico par que puede cruzar el corte es
         (ultimo de la izquierda, primero de la derecha) -> O(1).

    Recurrencia : T(N) = 2 T(N/2) + O(1)  ->  O(N) tras ordenar.
    Complejidad total : O(N log N) (por la ordenacion inicial).
    """
    ordenados = sorted(puntos)

    def resolver(sub: List[float]) -> Tuple[float, Tuple[float, float]]:
        # Caso base: 0 o 1 punto -> no hay par posible.
        if len(sub) < 2:
            return float('inf'), (None, None)
        # Caso base: exactamente 2 puntos.
        if len(sub) == 2:
            return distancia_1d(sub[0], sub[1]), (sub[0], sub[1])

        medio = len(sub) // 2
        dist_izq, par_izq = resolver(sub[:medio])
        dist_der, par_der = resolver(sub[medio:])

        # Mejor par dentro de cada mitad.
        if dist_izq <= dist_der:
            mejor_dist, mejor_par = dist_izq, par_izq
        else:
            mejor_dist, mejor_par = dist_der, par_der

        # Combinacion: par que cruza el corte (frontera entre mitades).
        dist_cruce = distancia_1d(sub[medio - 1], sub[medio])
        if dist_cruce < mejor_dist:
            mejor_dist = dist_cruce
            mejor_par = (sub[medio - 1], sub[medio])

        return mejor_dist, mejor_par

    return resolver(ordenados)
```

```python
# -----------------------------------------------------------------------------
#  EJEMPLO 1D : conjunto pequeno para ver el resultado paso a paso
# -----------------------------------------------------------------------------
ejemplo_1d = [3403, 4537, 9089, 9746, 7259, 4540]
print('Lista de puntos 1D:', ejemplo_1d)
print()

dist_fb, par_fb = mas_cercanos_1d_fuerza_bruta(ejemplo_1d)
dist_sc, par_sc = mas_cercanos_1d_orden_scan(ejemplo_1d)
dist_dv, par_dv = mas_cercanos_1d_divide_venceras(ejemplo_1d)

print(f'Fuerza Bruta        -> par {par_fb}  distancia = {dist_fb}')
print(f'Orden + Scan        -> par {par_sc}  distancia = {dist_sc}')
print(f'Divide y Venceras   -> par {par_dv}  distancia = {dist_dv}')
print()
print(f'Los dos puntos mas cercanos son {par_fb}, separados por {dist_fb} unidades.')
```

## Implementación en 2D

Se extiende el problema al plano cartesiano usando la **distancia euclidiana**. Se implementan dos enfoques:

1. **Fuerza Bruta** — $O(N^2)$
2. **Divide y Vencerás** — $O(N \log N)$

El paso de combinación filtra los puntos dentro de una **franja vertical** de ancho $2d$ alrededor de la mediana en X, y para cada punto solo compara con un número constante de vecinos (máximo 7).

```python
# =============================================================================
#  PROBLEMA EN 2D : Par de puntos mas cercanos en el plano
#  Punto = tupla (x, y)
# =============================================================================

# -----------------------------------------------------------------------------
#  1) FUERZA BRUTA 2D  ->  O(N^2)
# -----------------------------------------------------------------------------
def mas_cercanos_2d_fuerza_bruta(puntos: List[Tuple[float, float]]):
    """Compara la distancia euclidiana de todos los pares. O(N^2)."""
    n = len(puntos)
    mejor_dist = float('inf')
    mejor_par = (None, None)

    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_euclidiana(puntos[i], puntos[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (puntos[i], puntos[j])

    return mejor_dist, mejor_par


# -----------------------------------------------------------------------------
#  2) DIVIDE Y VENCERAS 2D  ->  O(N log N)
# -----------------------------------------------------------------------------
def _combinar_franja_2d(franja, d, mejor_par):
    """
    Revisa los pares candidatos dentro de la franja (ya ordenada por Y).
    Para cada punto basta comparar con los siguientes mientras la
    diferencia en Y sea menor que d (como maximo 7 vecinos). -> O(N).
    """
    mejor_dist = d
    m = len(franja)
    for i in range(m):
        j = i + 1
        while j < m and (franja[j][1] - franja[i][1]) < mejor_dist:
            dist = distancia_euclidiana(franja[i], franja[j])
            if dist < mejor_dist:
                mejor_dist = dist
                mejor_par = (franja[i], franja[j])
            j += 1
    return mejor_dist, mejor_par


def mas_cercanos_2d_divide_venceras(puntos: List[Tuple[float, float]]):
    """
    Algoritmo clasico Divide y Venceras en el plano:
      1. Pre-ordenar por X (Px) y por Y (Py).
      2. Dividir por la mediana en X.
      3. Resolver recursivamente cada mitad -> d = min(dL, dR).
      4. Combinar revisando una franja de ancho 2d en O(N).

    Recurrencia : T(N) = 2 T(N/2) + O(N)  ->  O(N log N).
    """
    px = sorted(puntos, key=lambda p: p[0])   # ordenado por X
    py = sorted(puntos, key=lambda p: p[1])   # ordenado por Y

    def resolver(px, py):
        n = len(px)
        # Caso base: pocos puntos -> fuerza bruta directa.
        if n <= 3:
            return mas_cercanos_2d_fuerza_bruta(px)

        medio = n // 2
        punto_medio = px[medio]

        izq_px = px[:medio]
        der_px = px[medio:]
        set_izq = set(map(id, izq_px))
        izq_py = [p for p in py if id(p) in set_izq]
        der_py = [p for p in py if id(p) not in set_izq]

        dist_izq, par_izq = resolver(izq_px, izq_py)
        dist_der, par_der = resolver(der_px, der_py)

        if dist_izq <= dist_der:
            d, mejor_par = dist_izq, par_izq
        else:
            d, mejor_par = dist_der, par_der

        # Franja de ancho 2d alrededor de la linea de division (ordenada por Y).
        franja = [p for p in py if abs(p[0] - punto_medio[0]) < d]
        return _combinar_franja_2d(franja, d, mejor_par)

    return resolver(px, py)
```

```python
# -----------------------------------------------------------------------------
#  EJEMPLO 2D : conjunto pequeno de puntos en el plano
# -----------------------------------------------------------------------------
ejemplo_2d = [(1122, 6175), (135, 4076), (7296, 2741), (1150, 6180), (8000, 9000)]
print('Lista de puntos 2D:', ejemplo_2d)
print()

dist_fb, par_fb = mas_cercanos_2d_fuerza_bruta(ejemplo_2d)
dist_dv, par_dv = mas_cercanos_2d_divide_venceras(ejemplo_2d)

print(f'Fuerza Bruta       -> par {par_fb}  distancia = {dist_fb:.4f}')
print(f'Divide y Venceras  -> par {par_dv}  distancia = {dist_dv:.4f}')
print()
print(f'Los dos puntos mas cercanos son {par_fb},')
print(f'separados por una distancia de {dist_fb:.4f} unidades.')
```

## Implementación en 3D

El algoritmo se generaliza al espacio tridimensional manteniendo la complejidad **$O(N \log N)$**.

1. **Fuerza Bruta** — $O(N^2)$
2. **Divide y Vencerás** — $O(N \log N)$

La división se hace con un **plano** $X = \text{mediana}$, y el paso de combinación filtra los puntos en una franja 3D de ancho $2d$. Por el argumento de empaquetamiento de esferas, cada punto solo necesita compararse con un número constante de vecinos (máximo 15).

```python
# =============================================================================
#  PROBLEMA EN 3D : Par de puntos mas cercanos en el espacio
#  Punto = tupla (x, y, z)
# =============================================================================

# -----------------------------------------------------------------------------
#  1) FUERZA BRUTA 3D  ->  O(N^2)
# -----------------------------------------------------------------------------
def mas_cercanos_3d_fuerza_bruta(puntos: List[Tuple[float, float, float]]):
    """Compara la distancia euclidiana 3D de todos los pares. O(N^2)."""
    n = len(puntos)
    mejor_dist = float('inf')
    mejor_par = (None, None)

    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_euclidiana(puntos[i], puntos[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (puntos[i], puntos[j])

    return mejor_dist, mejor_par


# -----------------------------------------------------------------------------
#  2) DIVIDE Y VENCERAS 3D  ->  O(N log N)
# -----------------------------------------------------------------------------
def _combinar_franja_3d(franja, d, mejor_par):
    """
    Franja ordenada por Y. Para cada punto recorremos los sucesores
    y nos detenemos cuando la diferencia en Y alcanza d. El numero de
    vecinos a comprobar esta acotado por una constante (<= 15). -> O(N).
    """
    mejor_dist = d
    m = len(franja)
    for i in range(m):
        j = i + 1
        while j < m and (franja[j][1] - franja[i][1]) < mejor_dist:
            dist = distancia_euclidiana(franja[i], franja[j])
            if dist < mejor_dist:
                mejor_dist = dist
                mejor_par = (franja[i], franja[j])
            j += 1
    return mejor_dist, mejor_par


def mas_cercanos_3d_divide_venceras(puntos: List[Tuple[float, float, float]]):
    """
    Divide y Venceras en 3D:
      1. Pre-ordenar por X y por Y.
      2. Dividir el espacio con el plano X = mediana.
      3. Resolver recursivamente -> d = min(dL, dR).
      4. Combinar dentro de una franja 3D de ancho 2d en O(N).

    Recurrencia : T(N) = 2 T(N/2) + O(N)  ->  O(N log N).
    """
    px = sorted(puntos, key=lambda p: p[0])   # ordenado por X
    py = sorted(puntos, key=lambda p: p[1])   # ordenado por Y

    def resolver(px, py):
        n = len(px)
        if n <= 3:
            return mas_cercanos_3d_fuerza_bruta(px)

        medio = n // 2
        punto_medio = px[medio]

        izq_px = px[:medio]
        der_px = px[medio:]
        set_izq = set(map(id, izq_px))
        izq_py = [p for p in py if id(p) in set_izq]
        der_py = [p for p in py if id(p) not in set_izq]

        dist_izq, par_izq = resolver(izq_px, izq_py)
        dist_der, par_der = resolver(der_px, der_py)

        if dist_izq <= dist_der:
            d, mejor_par = dist_izq, par_izq
        else:
            d, mejor_par = dist_der, par_der

        # Franja 3D: |p.x - mediana.x| < d  (mantenida ordenada por Y).
        franja = [p for p in py if abs(p[0] - punto_medio[0]) < d]
        return _combinar_franja_3d(franja, d, mejor_par)

    return resolver(px, py)
```

```python
# -----------------------------------------------------------------------------
#  EJEMPLO 3D : conjunto pequeno de puntos en el espacio
# -----------------------------------------------------------------------------
ejemplo_3d = [(1, 2, 3), (10, 10, 10), (2, 3, 4), (50, 40, 30), (3, 3, 3)]
print('Lista de puntos 3D:', ejemplo_3d)
print()

dist_fb, par_fb = mas_cercanos_3d_fuerza_bruta(ejemplo_3d)
dist_dv, par_dv = mas_cercanos_3d_divide_venceras(ejemplo_3d)

print(f'Fuerza Bruta       -> par {par_fb}  distancia = {dist_fb:.4f}')
print(f'Divide y Venceras  -> par {par_dv}  distancia = {dist_dv:.4f}')
print()
print(f'Los dos puntos mas cercanos son {par_fb},')
print(f'separados por una distancia de {dist_fb:.4f} unidades.')
```

```python
# =============================================================================
#  VERIFICACION DE CORRECTITUD Y BENCHMARK DE RENDIMIENTO
# =============================================================================

def cronometrar(funcion, *args):
    """Ejecuta una funcion y devuelve (resultado, tiempo_en_milisegundos)."""
    inicio = time.perf_counter()
    resultado = funcion(*args)
    fin = time.perf_counter()
    return resultado, (fin - inicio) * 1000.0


def generar_puntos(n, dim, rango=10000):
    """Genera n puntos aleatorios de la dimension indicada (1, 2 o 3)."""
    if dim == 1:
        return [random.uniform(0, rango) for _ in range(n)]
    return [tuple(random.uniform(0, rango) for _ in range(dim)) for _ in range(n)]


# -----------------------------------------------------------------------------
#  Parametros del experimento
# -----------------------------------------------------------------------------
random.seed(42)        # semilla fija -> resultados reproducibles
N = 500                # tamano del conjunto de puntos

datos_1d = generar_puntos(N, 1)
datos_2d = generar_puntos(N, 2)
datos_3d = generar_puntos(N, 3)

# -----------------------------------------------------------------------------
#  1) CORRECTITUD : todas las versiones deben dar la MISMA distancia minima
# -----------------------------------------------------------------------------
print('=' * 60)
print('VERIFICACION DE CORRECTITUD')
print('=' * 60)

(d1_fb, _), t1_fb = cronometrar(mas_cercanos_1d_fuerza_bruta, datos_1d)
(d1_sc, _), t1_sc = cronometrar(mas_cercanos_1d_orden_scan, datos_1d)
(d1_dv, _), t1_dv = cronometrar(mas_cercanos_1d_divide_venceras, datos_1d)
ok_1d = math.isclose(d1_fb, d1_sc) and math.isclose(d1_fb, d1_dv)
print(f'1D: {"OK" if ok_1d else "ERROR"}  (dist = {d1_fb:.4f})')

(d2_fb, _), t2_fb = cronometrar(mas_cercanos_2d_fuerza_bruta, datos_2d)
(d2_dv, _), t2_dv = cronometrar(mas_cercanos_2d_divide_venceras, datos_2d)
ok_2d = math.isclose(d2_fb, d2_dv)
print(f'2D: {"OK" if ok_2d else "ERROR"}  (dist = {d2_fb:.4f})')

(d3_fb, _), t3_fb = cronometrar(mas_cercanos_3d_fuerza_bruta, datos_3d)
(d3_dv, _), t3_dv = cronometrar(mas_cercanos_3d_divide_venceras, datos_3d)
ok_3d = math.isclose(d3_fb, d3_dv)
print(f'3D: {"OK" if ok_3d else "ERROR"}  (dist = {d3_fb:.4f})')

# -----------------------------------------------------------------------------
#  2) BENCHMARK : comparativa de tiempos (N = 500)
# -----------------------------------------------------------------------------
print()
print('=' * 60)
print(f'BENCHMARK DE RENDIMIENTO (N = {N})')
print('=' * 60)
print(f'{"Algoritmo":<28}{"Tiempo (ms)":>14}')
print('-' * 60)
print(f'{"1D Fuerza Bruta":<28}{t1_fb:>14.2f}')
print(f'{"1D Orden + Scan":<28}{t1_sc:>14.2f}')
print(f'{"1D Divide y Venceras":<28}{t1_dv:>14.2f}')
print(f'{"2D Fuerza Bruta":<28}{t2_fb:>14.2f}')
print(f'{"2D Divide y Venceras":<28}{t2_dv:>14.2f}')
print(f'{"3D Fuerza Bruta":<28}{t3_fb:>14.2f}')
print(f'{"3D Divide y Venceras":<28}{t3_dv:>14.2f}')

# -----------------------------------------------------------------------------
#  3) GRAN ESCALA : ventaja de Divide y Venceras en 2D (N = 5000)
# -----------------------------------------------------------------------------
print()
print('=' * 60)
print('GRAN ESCALA 2D (N = 5000)')
print('=' * 60)
datos_grande = generar_puntos(5000, 2)
_, t_fb_grande = cronometrar(mas_cercanos_2d_fuerza_bruta, datos_grande)
_, t_dv_grande = cronometrar(mas_cercanos_2d_divide_venceras, datos_grande)
print(f'Fuerza Bruta 2D      : {t_fb_grande/1000:>8.3f} s')
print(f'Divide y Venceras 2D : {t_dv_grande/1000:>8.3f} s')
print(f'Mejora de velocidad  : {t_fb_grande / t_dv_grande:>8.1f}x mas rapido')
```
